```python
%load_ext autoreload
%autoreload 2
import imageio
from robot_painting.pytorch_planar_scene_drawing import *
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from tqdm.notebook import tqdm
from IPython.display import display, HTML

import torch
cuda = torch.device('cuda')
```

```python
target_image_path = "../data/target.jpeg"
target_image_np = imageio.imread(target_image_path).astype(np.float)/256.
target_image = torch.tensor(target_image_np, device=cuda).permute(2, 0, 1)
n_channels, image_rows, image_cols = target_image.shape
b2p, p2b = oilpaint_converters(device=torch.device('cuda'))

print("Rows: ", image_rows)
print("Cols: ", image_cols)

brush_search_paths = ["../data/brushes/*.png"]
# Open every image in that folder
brush_paths = sum([glob.glob(sp) for sp in brush_search_paths], [])
print("Found brushes: ", brush_paths)

raw_brushes = [imageio.imread(brush_path).astype(np.float) / 256. for brush_path in brush_paths]
raw_brushes = numpy_images_to_torch(raw_brushes).cuda()
plt.imshow(raw_brushes[0, :].permute(1, 2, 0).cpu().numpy())
```

```python
torch.autograd.set_detect_anomaly(True)
def fit_brushstroke_with_grad_descent(current_image, target_image, brush, N_gd_iters=10, lr=1E-2):
    current_pre_oil = b2p(current_image)
    initial_error = torch.mean(torch.square(target_image[:3, ...] - current_image[:3, ...]))

    class StrokeRenderer(torch.nn.Module):
        def __init__(self, brush, make_plots=False):
            super().__init__()
            self.sprite_pose = torch.nn.Parameter(
                torch.tensor([
                    torch.randint(low=int(-image_rows / 2), high=int(image_rows / 2), size=(1,)),
                    torch.randint(low=int(-image_cols / 2), high=int(image_cols / 2), size=(1,)),
                    torch.rand(1) * np.pi * 2.
                ]))
            
            center = self.sprite_pose[0:2].detach()
            center_x = int(min(max(center[0].item() + (image_rows / 2), 0), image_rows))
            center_y = int(min(max(center[1].item() + (image_cols / 2), 0), image_cols))
            init_color = target_image[:, center_x, center_y]
            self.sprite_color_pre = torch.nn.Parameter(
                torch.logit(init_color)
            )
            self.sprite_scale_pre = torch.nn.Parameter(
                torch.rand(1)
            )
            self.brush = brush
            self.make_plots = make_plots

        def compute_loss(self):
            brush = self.brush.clone() # THIS SHOULDNT BE NECESSARY
            sprite_color = torch.sigmoid(self.sprite_color_pre)
            sprite_scale = torch.sigmoid(self.sprite_scale_pre)
            for i in range(3):
                brush[i, :, :] = sprite_color[i]

            sprite_ims = draw_sprites_at_poses(
                  self.sprite_pose.unsqueeze(0),
                  sprite_scale.unsqueeze(0),
                  brush.shape[1], brush.shape[2],
                  image_rows, image_cols,
                  brush.unsqueeze(0)
            )
            sprite_im_oil = b2p(sprite_ims)[0, ...]
            alpha = (1. - sprite_im_oil[3, :, :])
            new_image_oil = current_pre_oil[:3, :, :] * (1. - alpha) + sprite_im_oil[:3, :, :] * alpha
            new_image = p2b(new_image_oil)
            loss = torch.mean(torch.square(new_image[:3, ...] - target_image[:3, ...]))
            
            if self.make_plots:
                plt.figure()
                plt.subplot(1, 2, 1)
                plt.imshow(new_image.detach().permute(1, 2, 0).cpu().numpy())
                plt.subplot(1, 2, 2)
                plt.imshow(alpha.detach().cpu().numpy())

            return loss, new_image

    model = StrokeRenderer(brush)
    optimizer = torch.optim.Adam(
        model.parameters(),
        lr=lr
    )

    for iter_k in range(N_gd_iters):
        loss, new_image = model.compute_loss()
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    
    if loss < initial_error:
        return new_image.detach()
    else:
        print("Didn't improve error this time")
        return current_image.detach()

plt.figure()
plt.imshow(target_image.permute(1, 2, 0).cpu().numpy())
current_image = torch.ones_like(target_image).cuda()
images = [current_image.detach().permute(1, 2, 0).cpu().numpy().copy()]
for k in tqdm(range(2000)):
    current_image = fit_brushstroke_with_grad_descent(current_image, target_image, raw_brushes[0, ...].clone(), N_gd_iters=50, lr=0.1)
    images.append(current_image.detach().permute(1, 2, 0).cpu().numpy().copy())
plt.imshow(images[-1])
```

```python
dpi = 300
plt.figure(dpi=dpi).set_size_inches(image_rows/dpi, image_cols/dpi)
plt.axis('off')
plt.title("Recovered")
im_artists = []
for image in images[::100]:
    im_artists.append([plt.gca().imshow(image)])
ani = animation.ArtistAnimation(plt.gcf(), im_artists, interval=50, blit=True, repeat_delay=1000)
display(HTML(ani.to_html5_video()))
```
